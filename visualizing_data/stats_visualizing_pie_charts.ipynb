```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_bins = 5
total_n = 100

# a subtração por 0.01 garante que os números gerados serão sempre <= 5
raw_data = np.ceil( np.logspace(np.log10(1/2), np.log10(n_bins-.01), total_n) )
raw_data[:3]
```

```python
unique_nums = np.unique(raw_data)
data4pie = np.zeros(len(unique_nums))

for i in range(len(unique_nums)):
    data4pie[i] = sum(raw_data==unique_nums[i])
```

```python
data4pie
```

```python
plt.pie(data4pie, labels=100*data4pie/sum(data4pie))
plt.show()
```

```python
# another option
plt.pie(data4pie, labels=['zero', 'one', 'two', 'three', 'four'], explode=[0, .1, 0, .3, 0])
plt.show()
```

```python
## for continous data

# generate log-normal data
data = np.exp(np.random.randn(1000)/10)

histout = np.histogram(data, bins=6)
```

```python
plt.pie(histout[0], labels=100*histout[0]/sum(histout[0]))
plt.show()
```
