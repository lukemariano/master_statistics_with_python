```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# data sizes
m = 30 # rows
n = 6 # columns

data = np.zeros((m, n))

for i in range(n):
    data[:,i] = m*np.random.randn(m) * ( 2*i/(n-1) -1 )**2 + (i+1)**2
```

```python
data[:2]
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data.T)
plt.show()
```

```python
sns.boxplot(data)
plt.show()
```

```python
df = pd.DataFrame(data, columns=['zero', 'one', 'two', 'three', 'four', 'five'])
df.head()
```

```python
sns.boxplot(df, orient="h")
plt.show()
```

```python
%%markdown
### Concluses:
* Uma distribuio normal possui uma mediana extremamente prxima a mdia e assume que valores mais prximos a mdia tem uma maior probabilidade de ocorrncia. Possui uma cauda nos dados que representam valores com probabilidade baixa de ocorrer;
* Uma distribuio uniforme possui grfico mais achatado indicando uma probabilidade de ocorrncia de valores iguais para valores em um mesmo intervalo. No existem caudas porque todos os valores possuem a mesma probabilidade de ocorrncia;
```
