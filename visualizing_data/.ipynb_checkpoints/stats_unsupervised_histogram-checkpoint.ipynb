```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
n = 1000
k= 40

data = np.exp( np.random.randn(n)/2 )
data[:3]
```

```python
bins_count, bins_edges = np.histogram(data, k)
print("Densidades do histograma:", bins_count)
print("Bordas das caixas:", bins_edges)
```

```python
bins_height = sum(bins_count)
bins_relative = 100 * bins_count / bins_height
```

```python
bins_relative
```

```python
bins_centers = (bins_edges[:-1] + bins_edges[1:])/2
```

```python
plt.plot(bins_centers, bins_relative)
plt.show()
```

```python
plt.bar(bins_centers, bins_count, width=(bins_edges[1] - bins_edges[0]))
```

```python
k = 40

sns.histplot(data,bins=k)
plt.show()
```
