```python
x = np.linspace(0.1, 5, 100)

y1 = x
y2 = np.exp(x)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(x, y1, label=r"$x = x$", color="blue")
axes[0].plot(x, y2, label=r"$x = e^x$", color="cyan")
axes[0].set_title("Linear y-axis")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(x, y1, label=r"$x = x$", color="blue")
axes[1].plot(x, y2, label=r"$x = e^x$", color="cyan")
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic y-axis")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y (log scale)")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
