```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000

# log-normal distribution
data1 = np.exp( np.random.randn(n)/2 ) # altera o formato da distribuição
data2 = np.exp( np.random.randn(n)/10 ) # altera o formato da distribuição
data3 = np.exp( np.random.randn(n)/2 + 1 ) # altera a média da distribuição (soma uma unidade em todos os elementos do array)
```

```python
plt.plot(data1, 's')
plt.show()
```

```python
plt.plot(data2, 's')
plt.show()
```

```python
plt.plot(data3, 's')
plt.show()
```

```python
k = 20

plt.hist(data1, bins=k)
plt.hist(data2, bins=k)
plt.hist(data3, bins=k)

plt.show()
```

```python
y1, x1 = np.histogram(data1, bins=k)
xx1 = (x1[:-1] + x1[1:])/2
y1 = 100*y1 / sum(y1) # convert to probability / pct

y2, x2 = np.histogram(data2, bins=k)
xx2 = (x2[:-1] + x2[1:])/2
y2 = 100*y2 / sum(y2) # convert to probability / pct

y3, x3 = np.histogram(data3, bins=k)
xx3 = (x3[:-1] + x3[1:])/2
y3 = 100*y3 / sum(y3) # convert to probability / pct

plt.plot(xx1, y1, 's-', label='data1')
plt.plot(xx2, y2, 'o-', label='data2')
plt.plot(xx3, y3, '^-', label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```
