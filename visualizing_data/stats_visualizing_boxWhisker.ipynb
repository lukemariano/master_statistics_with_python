```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# data sizes
m = 30 # rows
n = 6 # columns

data = np.zeros((m, n))

for i in range(n):
    data[:,i] = m*np.random.randn(m) * ( 2*i/(n-1) -1 )**2 + (i+1)**2
```

```python
data[:2]
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data.T)
plt.show()
```

```python
sns.boxplot(data)
plt.show()
```

```python
df = pd.DataFrame(data, columns=['zero', 'one', 'two', 'three', 'four', 'five'])
df.head()
```

```python
sns.boxplot(df, orient="h")
plt.show()
```

```python
%%markdown
### Conclusões:
* Uma distribuição normal possui uma mediana extremamente próxima a média e assume que valores mais próximos a média tem uma maior probabilidade de ocorrência. Possui uma cauda nos dados que representam valores com probabilidade baixa de ocorrer;
* Uma distribuição uniforme possui gráfico mais achatado indicando uma probabilidade de ocorrência de valores iguais para valores em um mesmo intervalo. Não existem caudas porque todos os valores possuem a mesma probabilidade de ocorrência;
```
